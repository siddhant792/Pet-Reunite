```python
# Load model directly
from transformers import AutoImageProcessor, AutoModelForImageClassification

processor = AutoImageProcessor.from_pretrained("skyau/dog-breed-classifier-vit")
model = AutoModelForImageClassification.from_pretrained("skyau/dog-breed-classifier-vit")
```

```python
import torch
from PIL import Image
import requests
from io import BytesIO

image = 'https://upload.wikimedia.org/wikipedia/commons/9/99/Brooks_Chase_Ranger_of_Jolly_Dogs_Jack_Russell.jpg'
response = requests.get(image)
img = Image.open(BytesIO(response.content))


inputs = processor(img , return_tensors="pt")

with torch.no_grad():
    logits = model(**inputs).logits

# model predicts one of the 1000 ImageNet classes
predicted_label = logits.argmax(-1).item()
print(model.config.id2label[predicted_label])
```
